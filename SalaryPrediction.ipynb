```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("DigitalAd_dataset.csv")
```

```python
print(dataset.shape)
print(dataset.head(5))
```

```python
X = dataset.iloc[:,:-1].values
X
```

```python
Y = dataset.iloc[:,-1].values
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.10,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
# print(X_train)
print(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state=0)
model.fit(X_train,Y_train)
```

```python
age = int(input("Enter the Age of the User: "))
salary = int(input("Enter the Salary of the User: "))
newCus = [[age,salary]]
result = model.predict(sc.transform(newCus))
print(result)
if result == 1:
    print("User will Purchase the Product")
else:
    print("User will Not Purchase the product")
```

```python
y_pred = model.predict(X_test)
# print(np.concatenate((y_pred.reshape(len(y_pred),1), Y_test.reshape(len(Y_test),1),1)))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(Y_test,y_pred)
print("Confusion Matrix: ")
print(cm)

print("The Accuracy of the Model is: {0}%".format(accuracy_score(Y_test,y_pred)*100))

```
